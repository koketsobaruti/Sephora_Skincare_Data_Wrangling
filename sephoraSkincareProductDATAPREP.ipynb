```python
import pandas as pd

product_df = pd.read_csv('product_info.csv')

product_df.head(10)
```

```python
#get only the data that is skincare
skincare_df = product_df.loc[product_df['primary_category']=='Skincare']
skincare_df.reset_index(inplace=True)
skincare_df.drop(['index'], axis=1, inplace=True)
print('columns dropped')
skincare_df.head()
```

```python
#find out the number of rows
len(skincare_df)
```

```python
#check if there are any null columns
skincare_df.isnull().sum()
```

```python
#begin to drop the columns with too many NaN values
skincare_df.drop(['child_min_price','child_max_price','value_price_usd','sale_price_usd','variation_desc','child_max_price'], axis=1, inplace=True)
print('columns dropped')
```

```python
#check if there are any null columns
skincare_df.isnull().sum()
```

```python
#replace the values of some columns
skincare_df[['highlights','ingredients','size']]=skincare_df[['highlights','ingredients','size']].fillna('N/A')
```

```python
skincare_df[['variation_value']]=skincare_df[['variation_value']].fillna('N/A')
```

```python
#check if there are any null columns
skincare_df.isnull().sum()
```

```python
#drop the following coolumn because it is redundant
skincare_df.drop(['variation_type'], axis=1, inplace=True)
print('columns dropped')
```

```python
#replace the values of some columns
skincare_df['tertiary_category']=skincare_df['tertiary_category'].fillna('None')
```

```python
#check if there are any null columns
skincare_df.isnull().sum()
```

```python
skincare_df[['rating','reviews']]=skincare_df[['rating','reviews']].fillna('0.0')
```

```python
skincare_df[['rating','reviews']] = skincare_df[['rating','reviews']].astype(float)
```

```python
#find out the datatypes
skincare_df.dtypes
```

```python
#normalise the columns for loves_count using min-max scaling 
skincare_df["loves_count"] = skincare_df['loves_count']/skincare_df['loves_count'].max()
```
